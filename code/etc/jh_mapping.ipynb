```python
from pyproj import Proj, transform
```

```python
sports_center = pd.read_csv("/content/drive/Shareddrives/강서구/220304 mapping/서울특별시강서구체육도장업인허가정보.csv", encoding="CP949", skiprows = [0], on_bad_lines='skip')
sports_center = sports_center.dropna(subset=['x', 'y'])
sports_center = sports_center[sports_center['dtlstatenm']=='영업중']
sports_center = sports_center.reset_index(drop=True)
```

```python
proj_UTMK = Proj(init='epsg:2097') 
proj_WGS84 = Proj(init='epsg:4326') 


def transform_utmk_to_w84(df):
    return pd.Series(transform(proj_UTMK, proj_WGS84, df['x'], df['y']), index=['x', 'y'])

sports_center[['x', 'y']] = sports_center.apply(transform_utmk_to_w84, axis=1)
```

```python
# 서울특별시강서구청소년게임제공업인허가정보
game = pd.read_excel('/content/drive/Shareddrives/강서구/220304 mapping/서울특별시강서구청소년게임제공업인허가정보.xls', skiprows=1)
game = game[game['dtlstatenm']=='영업중']
game = game.reset_index(drop=True)
proj_UTMK = Proj(init='epsg:2097') 
proj_WGS84 = Proj(init='epsg:4326')
def transform_utmk_to_w84(df):
    return pd.Series(transform(proj_UTMK, proj_WGS84, df['x'], df['y']), index=['x', 'y'])
game[['x', 'y']] = game.apply(transform_utmk_to_w84, axis=1)
game_map = folium.Map(location=[game['y'][0], game['x'][0]], zoom_start=12)
for i in range(len(game)):
    folium.Circle(
        location=[game['y'][i], game['x'][i]],
        radius=30, # 원 크기
        color='tomato', # 원 선 색상
        fill_color='tomato', # 원 내부 색상
        popup='Circle popup',
        tooltip='Circle tooltip'
    ).add_to(game_map)
game_map.add_child(folium.LatLngPopup())
game_map
```
